```python
import os
import googleapiclient.discovery
import googleapiclient.errors
# from polyglot.detect import Detector
import re
import fasttext
# from huggingface_hub import hf_hub_download
import warnings
from google.colab import userdata
from ftlangdetect import detect
from googleapiclient.discovery import build
API=userdata.get('Yt_API')
```

```python
from googletrans import Translator
from google.colab import userdata
import nltk
import warnings
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
nltk.download('vader_lexicon')
def get_video_info(api_key, video_id):
    youtube = build('youtube', 'v3', developerKey=api_key)
    request = youtube.videos().list(
        part="snippet,statistics",
        id=video_id
    )
    response = request.execute()

    if response['items']:
        video_info = response['items'][0]
        channel_name = video_info['snippet']['channelTitle']
        likes = video_info['statistics']['likeCount']
        views = video_info['statistics']['viewCount']
        video_title = video_info['snippet']['title']
        return channel_name, likes, views, video_title
    else:
        return None
def clean_comment(text):
    # Define a regular expression to match HTML tags
    tag_pattern = re.compile(r'<[^>]+>')

    # Define a regular expression to match URLs

    url_pattern = re.compile(r'https?://\S+|www\.\S+')

    # Remove HTML tags and URLs from the text
    text_without_tags_and_links = re.sub(tag_pattern, '', text)
    text_without_tags_and_links = re.sub(url_pattern, '', text_without_tags_and_links)

    return text_without_tags_and_links

# def detect_language(text):
#     with warnings.catch_warnings():
#       # Suppress the specific warning from Polyglot
#       warnings.filterwarnings("ignore", category=UserWarning, module="polyglot")
#     try:
#         detector = Detector(text)
#         language = detector.language.name
#         # confidence = detector.language.confidence
#         return language
#     except Exception as e:
#         return "emoji"
def detect_language(text):
    warnings.filterwarnings("ignore", category=UserWarning)
    # model_path = hf_hub_download(repo_id="facebook/fasttext-language-identification", filename="model.bin")
    # warnings.filterwarnings("ignore")
    try:
      # model = fasttext.load_model(model_path)
      # # model.predict("Hello, world!")
      # t=model.predict(text, k=1)
      # return t
      return detect(text=text,low_memory=True)
    except Exception as e:
      print(text)
      return 'emoji'
def return_comments(DEVELOPER_KEY,videoId,n):
    api_service_name = "youtube"
    api_version = "v3"


    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey=DEVELOPER_KEY)

    request = youtube.commentThreads().list(
        part="snippet",
        videoId=videoId,
        maxResults=n
    )
    response = request.execute()
    comments=[]
    for item in response['items']:
        comments.append(item['snippet']['topLevelComment']['snippet']['textDisplay'])
    return comments
def translate_text(text, target_language='en'):
    translator = Translator()
    translation = translator.translate(text, dest=target_language)
    return translation.text
# def detect_language(text):
#     try:
#         language = detect(text)
#         return language
#     except Exception as e:
#         return 'emoji'
def analyze_sentiment(text):
    # Create a SentimentIntensityAnalyzer object
    #nltk.download('vader_lexicon')
    warnings.filterwarnings("ignore")
    sia = SentimentIntensityAnalyzer()

    # Get sentiment scores for the text
    sentiment_scores = sia.polarity_scores(text)

    # Determine the sentiment based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        sentiment = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'

    return sentiment, sentiment_scores
def return_df(videoId, n):
    comments = return_comments(API, videoId, n)
    print('Comments received...')

    pred = []
    print('Thinking...')
    lan = []
    channel_name, likes, views, video_title = get_video_info(API, videoId)
    for comment in comments:
        comment = clean_comment(comment)
        if comment=='':
          comment='hi'
        temp=detect_language(comment)
        lan.append(languages[temp['lang']])
        comment = translate_text(comment)
        if detect_language(comment) == 'emoji':
            pass
        pred.append(analyze_sentiment(comment))
    print(pred)
    label = []
    compound = []
    for i in pred:
        label.append(i[0])
        compound.append(i[1]['compound'])

    data = pd.DataFrame({ 'Label': label, 'Language': lan, 'CompundScore': compound,'Comment': comments})

    return channel_name, likes, views, video_title,data
```

```python
import dash
from dash import dcc, html, dash_table
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
from wordcloud import WordCloud

# Initialize Dash app
app = dash.Dash(__name__)

# Layout of the web application
app.layout = html.Div(children=[
    dcc.Input(id='video-id-input', type='text', value=''),
    dcc.Input(id='n-input', type='number', value=1),
    html.Div(id='output-div'),

    dcc.Dropdown(id='label-dropdown'),

    dcc.Loading(
        id="loading",
        type="circle",
        children=[
            dash_table.DataTable(id='data-table', columns=[], data=[],page_size=10),  # Limiting to 10 rows per page
            dcc.Graph(id='word-cloud'),
            # Distribution of Compound Scores
            dcc.Graph(id='compound-scores-histogram'),

            # Count plot of Labels
            dcc.Graph(id='label-count-plot'),

            # Language distribution
            dcc.Graph(id='language-distribution'),

            # Boxplot for Compound Scores by Label
            dcc.Graph(id='boxplot-compound-scores'),

            # Scatter plot of Compound Scores by Comment Index
            dcc.Graph(id='scatter-plot-compound-scores'),

            # Pie chart for Language distribution
            dcc.Graph(id='pie-chart-languages')

        ]
    )
])

# Callback to update graphs and dropdown based on user input
@app.callback(
    [Output('output-div', 'children'),
     Output('word-cloud', 'figure'),
     Output('compound-scores-histogram', 'figure'),
     Output('label-count-plot', 'figure'),
     Output('language-distribution', 'figure'),
     Output('boxplot-compound-scores', 'figure'),
     Output('scatter-plot-compound-scores', 'figure'),
     Output('pie-chart-languages', 'figure'),
     Output('data-table', 'columns'),
     Output('data-table', 'data'),
     Output('label-dropdown', 'options'),
     Output('label-dropdown', 'value')],
    [Input('video-id-input', 'value'),
     Input('n-input', 'value'),
     Input('label-dropdown', 'value')]
)
def update_charts(video_id, n, selected_label):
    if video_id and n:
        try:
            channel_name, likes, views, video_title,comments_df = return_df(video_id, n)
            label_options = [{'label': label, 'value': label} for label in comments_df['Label'].unique()]
            selected_label = selected_label or label_options[0]['value']
            label_df = comments_df[comments_df['Label'] == selected_label]
            wordcloud = WordCloud(width=800, height=400, background_color='white').generate(" ".join(comments_df['Comment']))
            wordcloud_fig = px.imshow(wordcloud.to_array(), title=f'Word Cloud for Label {selected_label}')
            histogram_fig = px.histogram(comments_df, x='CompundScore', nbins=30, title='Distribution of Compound Scores')
            label_count_fig = px.bar(comments_df['Label'].value_counts(), x=comments_df['Label'].value_counts().index,y='Label', labels={'Label': 'Count'}, title='Distribution of Comment Labels')
            language_distribution_fig = px.pie(comments_df['Language'].value_counts(),values=comments_df['Language'].value_counts().values,names=comments_df['Language'].value_counts().index,title='Distribution of Languages')
            boxplot_fig = px.box(comments_df, x='Label', y='CompundScore', title='Compound Scores by Label')
            scatter_fig = px.scatter(comments_df, x=comments_df.index, y='CompundScore', title='Compound Scores by Comment Index')
            pie_chart_languages_fig = px.pie(comments_df['Language'].value_counts(),names=comments_df['Language'].value_counts().index,title=f'Pie Chart of Languages for {selected_label }')

            columns = [{'name': col, 'id': col} for col in comments_df.columns]
            data = comments_df.head(10).to_dict('records')  # Displaying only first 10 rows

            return (
                html.Div([
                    html.P(f'You entered: Video ID - {video_id}'),
                    html.P(f'n - {n}'),
                    html.P(f'Selected Label - {selected_label}'),
                    html.P(f'The channel name is {channel_name}'),
                    html.P(f'Video title :{video_title}')
                ]),
                wordcloud_fig,
                histogram_fig,
                label_count_fig,
                language_distribution_fig,
                boxplot_fig,
                scatter_fig,
                pie_chart_languages_fig,
                columns,
                data,
                label_options,
                selected_label
            )
        except Exception as e:
            error_message = f"An error occurred: {str(e)}"
            print(error_message)
            return error_message, {},{}, {}, {}, {}, {}, {}, [], [], [], None
    else:
        return "", {},{}, {}, {}, {}, {}, {}, [], [], [], None


# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
